```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_parquet('.\data\Crypto\BTCUSDT\YEAR_AND_MONTH=202305.parquet')
```

```python
print(df)
```

```python
import matplotlib.pyplot as plt

# Vẽ biểu đồ đường
plt.plot(df.index, df['OPEN'])
plt.plot(df.index, df['CLOSE'])
plt.xlabel('OPEN_TIME')
plt.ylabel('CLOSE')
plt.title('Price Trend')
plt.xticks(rotation=45)
plt.show()
```
